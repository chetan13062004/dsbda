```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("E:/PICT 3rd/subject material/DSBDA Prep/Datasets/PICT Datasets/heart.csv")
```

```python

# Histogram of Age
plt.figure()
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
# Pie chart for Heart Disease target
labels = ['No Disease', 'Disease']
sizes = df['target'].value_counts()

plt.figure()
plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['lightcoral', 'lightgreen'])
plt.title('Heart Disease Distribution')
plt.show()
```

```python
#box plot
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='target',y='age',data=df)
plt.show()
```

```python
#scatter plot
plt.figure()
sns.scatterplot(x='age', y='chol', data=df, hue='target')
plt.title('Age vs Cholesterol by Heart Disease')
plt.show()
```

```python
plt.figure()
sns.boxplot(x='target', y='chol', data=df)
sns.stripplot(x='target', y='chol', data=df, color='red', jitter=True)
plt.title('Cholesterol by Heart Disease (Box + Scatter)')
plt.show()
```
