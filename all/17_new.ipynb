```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("E:/PICT 3rd/subject material/DSBDA Prep/Datasets/PICT Datasets/heart.csv")
```

```python
df.head()
```

```python
#histogram
```

```python
df.hist(column=["age","chol","thalach"],figsize=(10,6),color="yellow",edgecolor="black")
plt.suptitle("Histogram")
plt.tight_layout()
plt.show()
```

```python
#dot plot
```

```python
plt.figure(figsize=(8,6))
sns.stripplot(x="target",y="age",data=df,jitter=True,hue="target",palette="Set2",legend=True)
plt.title("Box plot")
plt.xlabel("heart diasess")
plt.ylabel("age")
plt.show()
```

```python
# c. Bar Plot
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x="cp",data=df,hue=True,palette="Set2")
plt.xlabel("CP")
plt.ylabel("count")
plt.show()
```

```python
# d. Line Chart
```

```python
avg_chol=df.groupby("age")['chol'].mean().reset_index()
plt.figure(figsize=(8,6))
plt.plot(avg_chol['age'],avg_chol['chol'],marker='o',linestyle='-',color='red')
plt.grid(True)
plt.show()
```

```python
# Box + Histogram (age)
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='target',y='age',data=df,palette='Set2')
sns.histplot(x='age',data=df,bins=15,color='red')
plt.show()
```
