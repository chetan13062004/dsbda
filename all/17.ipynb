```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("E:/PICT 3rd/subject material/DSBDA Prep/Datasets/PICT Datasets/heart.csv")
```

```python
df.head()
```

```python
#histogram
```

```python
df.hist(column=['age','chol','thalach','trestbps'],figsize=(10,6),color='skyblue',edgecolor='black')
plt.suptitle('Histogram of Age, Cholesterol, Max Heart Rate, and Resting BP')
plt.tight_layout()
plt.show()
```

```python
#dot plot
```

```python
plt.figure(figsize=(8,5))
sns.stripplot(x='target', y='age', data=df, jitter=True, hue='target', palette='Set2', legend=True)
plt.title('Dot Plot of Age by Heart Disease Target')
plt.xlabel('Heart Disease (1 = Yes, 0 = No)')
plt.ylabel('Age')
plt.show()
```

```python
# c. Bar Plot
# Objective: To compare the frequency of chest pain types across patients
plt.figure(figsize=(6, 4))
sns.countplot(x='cp', data=df,hue=True,palette='Set2')
plt.title('Bar Plot of Chest Pain Types')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.show()
```

```python
# d. Line Chart
# Objective: To visualize trends - simulate average cholesterol across different ages
avg_chol_by_age = df.groupby('age')['chol'].mean().reset_index()
plt.figure(figsize=(8, 5))
plt.plot(avg_chol_by_age['age'], avg_chol_by_age['chol'], marker='o', linestyle='-', color='green')
plt.title('Line Chart: Average Cholesterol by Age')
plt.xlabel('Age')
plt.ylabel('Average Cholesterol')
plt.grid(True)
plt.show()
```

```python
# e. Box Plot with Histogram and Scatter (composite visualization)
# Objective: To visualize data distribution, outliers, and relationships

# Box + Histogram (age)
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='age', data=df, palette='Set3')
sns.histplot(data=df, x='age', bins=20, kde=True, color='lightblue')
plt.title('Box Plot with Histogram: Age and Heart Disease')
plt.show()

# Box + Scatter (chol)
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='chol', data=df, palette='cool')
sns.stripplot(x='target', y='chol', data=df, color='red', alpha=0.5, jitter=True)
plt.title('Box Plot with Scatter: Cholesterol and Heart Disease')
plt.show()
```
