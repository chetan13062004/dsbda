```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("E:/PICT 3rd/subject material/DSBDA Prep/Datasets/PICT Datasets/adult_dataset.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df['age'],bins=20,kde=True,color='yellow')
plt.show()
```

```python
#dot 
```

```python
plt.figure(figsize=(15,8))
sns.stripplot(x='education',y='educational-num',hue='income',data=df,jitter=True)
plt.show()
```

```python
#bar
```

```python
plt.figure(figsize=(15,6))
sns.barplot(x='occupation',y='hours-per-week',data=df, estimator='mean')
plt.show()
```

```python
#line chart
```

```python
edu_df=df.groupby('educational-num')['age'].mean().reset_index()
```

```python
plt.figure(figsize=(15,6))
plt.plot(edu_df['educational-num'],edu_df['age'],color='red')
plt.grid()
plt.show()
```

```python
#histo + sctter
```

```python
plt.figure(figsize=(15,6))
sns.histplot(df['age'], bins=20, kde=True)
sns.scatterplot(x=range(len(df)), y=df['age'], alpha=0.5)
plt.title("Age Distribution: Histogram + Scatter Overlay")
plt.xlabel("Index")
plt.ylabel("Age")
plt.show()
```
